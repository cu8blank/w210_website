```python
import pandas as pd
import numpy as np
import csv
from dateutil.parser import parse
```

```python
# Currency Predictions
col_curr = ["currency","rate_curr","curr_date","rate_7","rate_30","rate_90"]
df_curr_pred = pd.DataFrame(columns = col_curr)

curr_date = parse("April 24, 2016")

with open('April_24_Currency_Predictions.csv', 'rb') as csvfile:
    csv_val = csv.reader(csvfile)
    for row in csv_val:
        #Creates a numpy array
        new_curr = np.array([row[0].strip(),row[1],curr_date, row[2],row[3],row[4]])
        #Appends array to data frame
        df_curr_pred = df_curr_pred.append(pd.DataFrame(new_curr, index=col_curr).transpose())  

print df_curr_pred.head(10)
```

```python
# Cost of Living Index (used to be Big Mac Data)
col_big_mac = ["country","country_code","date","price"]
df_big_mac = pd.DataFrame(columns = col_big_mac)

with open('cost_living_data_parsed_plus_code.csv', 'rb') as csvfile:
    csv_val = csv.reader(csvfile)
    for row in csv_val:
        
        if (parse(row[2]).year) < 2016:
            break

        #Creates a numpy array
        new_big_mac = np.array([row[0],row[1],row[2],row[3]])
        #Appends array to data frame
        df_big_mac = df_big_mac.append(pd.DataFrame(new_big_mac, index=col_big_mac).transpose())

print df_big_mac.head(10)
```

```python
# Currency Country Data
col_country = ["currency_key","missing_curr","curr","country_code","country"]
df_country = pd.DataFrame(columns = col_country)

with open('Currency_Country.csv', 'rb') as csvfile:
    csv_val = csv.reader(csvfile)
    for row in csv_val:

        #Creates a numpy array
        new_country = np.array([row[0].strip(),row[1],row[2],row[3],row[4]])
        #Appends array to data frame
        df_country = df_country.append(pd.DataFrame(new_country, index=col_country).transpose())
        
print df_country.head(10)
```

```python
# map all three into a single flat file

df_big_country = pd.merge(left=df_country,right=df_big_mac, how='left', left_on='country_code', right_on='country_code')

df_big_country = df_big_country[0:174]
df_big_country.tail(10)
```

```python
# map all three into a single flat file

df_big_country_curr = pd.merge(left=df_big_country,right=df_curr_pred, how='left', left_on='currency_key', right_on='currency')

df_big_country_curr
```
